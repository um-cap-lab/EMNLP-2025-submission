```python


def clean_switchboard(filename):
    f = open(filename, "r")
    lines = f.readlines()
    cleaned_lines = []
    for line in lines:
        # Remove the first two characters from each line
        clean = line[2:]
        # remove leading spaces
        clean = clean.lstrip()
        #save the cleaned lines to a new file
        cleaned_lines.append(clean)
    f.close()
    new_filename = filename.replace(".train", "_cleaned.train")
    with open(new_filename, "w") as f:
        for line in cleaned_lines:
            f.write(line)

filename = "train_100M/switchboard.train"
clean_switchboard(filename)




# filename = "train_10M/switchboard.train"
# f = open(filename, "r")
# lines = f.readlines()

# cleaned_lines = []
# for line in lines:
#     # Remove the first two characters from each line
#     clean = line[2:]
#     # remove leading spaces
#     clean = clean.lstrip()
#     #save the cleaned lines to a new file
#     cleaned_lines.append(clean)
# print(f"Number of lines after removing empty and header lines: {len(cleaned_lines)}")
    

# f.close()
# #save the cleaned lines to a new file
# with open("train_10M/switchboard_cleaned.train", "w") as f:
#     for line in cleaned_lines:
#         f.write(line)
```

```python
def clean_open_subtitles(filename):
    f = open(filename, "r")
    lines = f.readlines()

    cleaned_lines = []
    for line in lines:
        if line.startswith("-"):
            cleaned_lines.append(line[1:])
        elif line.startswith("*"):
            clean = line.strip("*")
            clean = clean.lstrip()
            cleaned_lines.append(clean)
        else:
            cleaned_lines.append(line)    
    f.close()
    new_filename = filename.replace(".train", "_cleaned.train")
    with open(new_filename, "w") as f:
        for line in cleaned_lines:
            f.write(line)

filename100 = "train_100M/open_subtitles.train"
filename10 = "train_10M/open_subtitles.train"
clean_open_subtitles(filename10)
clean_open_subtitles(filename100)
```
